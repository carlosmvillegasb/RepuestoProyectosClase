### Respuesta [1]

Primero definimos una función llamada bisección que tome como argumentos a la función $f$ cuyas raices queremos encontrar, y a los extremos $a$ y $b$ del intervalo. Como queremos que se ingresen sólo funciones en el primer argumento, y sólo números reales en los otros dos (porque el método de bisección aplica a cálculo de variable real, que es en donde el teorema del valor intermedio tiene validez), usamos la estructura:


function biseccion(f::Function,a::Real,b::Real)

```julia
function biseccion(f::Function,A::Real,B::Real)
    tolerancia=0.00000000001
    if abs(A-B)<=tolerancia
        error("intervalo no válido")
    elseif A>B
        a=B
        b=A
    else
        a=A
        b=B
    end
    if f(a)*f(b)>0
        error("No puede garantizarse que haya raíces")
        elseif abs(f(a))<=tolerancia
        c=a
        elseif abs(f(b))<=tolerancia
        c=b
    elseif f(a)*f(b)<0
        c=(a+b)/2
    end
    if abs(f(c))<=tolerancia
        c
        elseif f(a)*f(c)<0
        biseccion(f,a,c)
        elseif f(b)*f(c)<0
        biseccion(f,c,b)
    else
        error("error inesperado")
    end
end
```

En la celda anterior lo que hicimos fue usar tres series de puertas condicionales sucesivas para ejecutar el algoritmo. La primera de ellas toma los argumentos reales (llamados A y B). Si A y B son iguales (representado por el que el valor absoluto de su diferencia sea menos que la tolerancia del algoritmo), regresa un error para decir que el intervalo no es válido. De lo contrario, si A es mayor que B, asigna a A como el valor de $b$ y a B como el valor de $a$, y vice-versa si A es menor que B. En otras palabras, esta primera puerta construye el intervalo $[a,\,b]$ asignando como $a$ el menor de los argumentos reales de la función y como $b$ al mayor de éstos. 

La siguiente puerta se encarga de ver si hay raices o no en el intervalo $[a,\,b]$ basados en el teorema del valor medio. Si no se satisface la hipótesis del teorema, regresa un error. Si la función $f$ evaluada en alguno de los extremos del intervalo da un número con valor absoluto menor a la tolerancia, el algoritmo lo toma como la raíz de la función. De lo contrario, calcula el punto medio $c$ para empezar el método de bisección.

A partir de ahí, la tercera puerta es para aplicar recursivamente el método de bisección. Se compara el signo de $f(c)$ con el de $f(a)$ y el de $f(b)$ para ver en qué intervalo está la raíz ir repetir el proceso de sacar el punto medio, hasta que el valor absoluto de $f(c)$ sea menor a la tolerancia para tomar a esa $c$ como la raíz. A pesar de que una vez que se tiene que $f(a)f(b)<0$ ($f(a)$ y $f(b)$ son de signo opuesto), $f(c)$ debe tener un signo distinto a al menos uno de ellos (o ser cero, en cuyo caso $c$ sería la raíz), se implementó un mensaje de error que apareciera si no se cumple ninguna condición (porque algo debería pasar inesperadamente mal para que eso ocurriera). 

Ahora hemos de probar esta función para hallar raíces de funciones conocidas (como polinomios y trigonométricas). Para no tener que estarlas definiendo aparte (por ahorita), simplemente escribimos x->f(x) en el primer argumento de la función bisección, donde f representa a la función $f$ cuya raíz vamos a calcular.

La posición de la partícula libre está dada por:

$$x(t)=x_{0}+vt,$$

donde $x_{0}$ es la posición inicial al tiempo $t=0$.

Mientras tanto, la posición del oscilador armónico simple está dada por:

$$x(t)=x_{e}+A\cos(\omega t+\phi),$$

donde $x_{e}$ es la posición de equilibrio entorno a la cuál oscila la partícula y $\phi$ es una fase que determina su posición inicial al tiempo $t=0$.

#### Inciso (a)

```julia
posicion_libre(x0::Real,v::Real,t::Real)=x0+v*t
```

```julia
posicion_oscilador(A::Real,frec::Real,t::Real,xe::Real=0,fase::Real=0)=xe+A*cos(frec*t+fase)
```

#### Inciso (c)

Ahora creamos una función colision que calcule el tiempo de colisión $\tau$ entre ambas partículas. A pesar de que esta función debe aceptar como argumentos a los parámetros relevantes del sistema, para poder usar la función bisección implementada en la respuesta [1], debemos definir dentro de colision una función $f$ de una sola variable (el tiempo) que pueda ser el argumento de la función biseccion. Dicha función también debe ser capaz de calcular el intervalo $[a,\,b]$ donde pueda hallarse $\tau$. En este caso $a$ es el tiempo en el que la partícula libre está en $x_{e}-A$ y $b$ es el tiempo donde ésta está en $x_{e}+A$.

$$a=\frac{x_{e}-A-x_{0}}{v}$$
$$b=\frac{x_{e}+A-x_{0}}{v}$$

Como estamos dividiendo entre $v$, es importante que la velocidad sea distinta de cero. Si la velocidad es cero, simplemente usamos otro método para calcular el tiempo de colisión, que es cuando el oscilador pase por $x_{0}$. En ese caso, el único intervalo que tiene sentido es uno entre el cuál el oscilador llegue a sus dos extremos, lo cuál ocurre cuando $\omega t +\phi$ es $0$ o $\pi$:

$$a=\frac{-\phi}{\omega}$$
$$b=\frac{\pi-\phi}{\omega}$$

Como se divide entre $\omega$, ésta también debe ser distinta de cero. Sin embargo, si ya estamos en el caso donde tanto $v$ como $\omega$ son cero, es porque las partículas están fijas en un punto y no están moviéndose, por lo que no pueden chocar. En dado caso, la función manda el mensaje de error apropiado.

```julia
function colision(x0_libre::Real,velocidad::Real,Amplitud::Real,frecuencia::Real,equilibrio::Real=0,fase::Real=0)
    func(t)=posicion_libre(x0_libre,velocidad,t)-posicion_oscilador(Amplitud,frecuencia,t,equilibrio,fase)
    N=100
    contador=0
    if velocidad!=0
        a=(equilibrio-Amplitud-x0_libre)/velocidad
        b=(equilibrio+Amplitud-x0_libre)/velocidad
        elseif frecuencua!=0
        a=(-fase)/frecuencia
        b=(pi-fase)/frecuencia
    else
        error("Las partículas no se mueven")
    end
    h=(b-a)/N
    for n in 0:(N-1)
        an=a+n*h
        bn=an+h
        if func(an)*func(bn)<0 && an>0
            contador=contador+1
        println(biseccion(func,an,bn))
        end
    end
    if contador==0
        println("No se encontraron colisiones")
    end
end
```

Aquí, después de calcular $a$ y $b$, se parte el intervalo $[a,\,b]$ en $N$ subintervalos  $[a_{n},\,b_{n}]$. En cada uno, la función colisión verifica si hay raices en ellos. Si sí, aplica la función biseccion, y si no, se va al siguiente subintervalo. De ese modo, la función biseccion sólo devuelve números, los cuales van siendo impresos con println(). Además, se implementó una variable contador que registra cuantas raices (colisiones) se hallaron en todos los subintervalos. Si no hubo ninguna en toda la ejecución de la función colision, ésta regresa un mensaje apropiado.

Esta respuesta coincide con las tres intersecciones de la gráfica del Inciso (a), por lo que nuestra función colision está bien implementada.

#### Inciso (d) [Opcional]

Como dijimos en el último párrafo, sólo tomaremos como posibles colisiones a las intersecciones entre las gráficas de las funciones de posición de ambas partículas para $t>0$. Además, ahora hemos de encontrar la verdadera colisión, que es la primera de esas intersecciones. Para ello, simplemente tomamos la función colision y le añadimos otra condición en el algoritmo. Dicha condición consta en declarar una variable fill una vez que se construyó el intervalo $[a,\,b]$, y se inicializa a fill en un valor que quede fuera y que sea mayor a $b$ (en este caso, $b+1$). Se modificó el algoritmo de tal forma que sólo cuente las raíces positivas de $f$ (si las hay, que es cosa de que se satisfagan $a_n>0$ y $f(a_{n})f(b_{n})<0$ en el correspondiente intervalo $[a_{n},\,b_{n}]$) si el valor de fill es mayor a $b$. Cuando se cumplen esas tres condiciones por primera vez, es porque el programa dio con la primera raíz de $f$. En ese caso, se incrementa el contador y se reasigna el valor de fill a aquél encontrado con el método de bisección. Como a partir de ese punto fill será menor que $b$, ya no se volverá a cambiar éste y tampoco seguirá el contador. Al final, la función regresa el valor de fill. (Sin embargo, si nunca se cambiaron el contador ni el valor de fill, el algoritmo transforma a fill en un string que da el mensaje de que no se encotraron colisiones). 

```julia
function colision_unica(x0_libre::Real,velocidad::Real,Amplitud::Real,frecuencia::Real,equilibrio::Real=0,fase::Real=0)
    func(t)=posicion_libre(x0_libre,velocidad,t)-posicion_oscilador(Amplitud,frecuencia,t,equilibrio,fase)
    N=100
    contador=0
    if velocidad!=0
        a=(equilibrio-Amplitud-x0_libre)/velocidad
        b=(equilibrio+Amplitud-x0_libre)/velocidad
        elseif frecuencua!=0
        a=(-fase)/frecuencia
        b=(pi-fase)/frecuencia
    else
        error("Las partículas no se mueven")
    end
    h=(b-a)/N
    fill=b+1
    for n in 0:(N-1)
        an=a+n*h
        bn=an+h
        if func(an)*func(bn)<0 && an>0 && fill>b
            contador=contador+1
            fill=biseccion(func,an,bn)
        end
    end
    if contador==0
        fill="No se encontraron colisiones"
    end
    return fill
end
```

Para verificar su funcionamiento, la evaluamos en los mismos valores de los argumentos que el inciso anterior:

```julia
colision_unica(-5,1,3,1)
```

```julia
colision_unica(-5,-1,3,1)
```

Notamos que ahora la salida de la función colision_unica es un float cuando sí hay colisión y no da el tiempo de dicha colisión. El hecho de que nos dé ese tiempo como un número es bastante útil si queremos meter esa función en otras más complicadas. Y cuando no hay colisión, devuelve un string.

### Respuesta [3]

El método de Newton sirve para encontrar una raíz de una función $f$ usando información de su derivada $f'$, partiendo de un valor inicial $x_{0}$ (que es una adivinanza) y obteniendo una sucesión de valores $x_{n}$ que convergen a la raíz buscada. 

Dado un punto $x_{n}$ de esa sucesión, se obtiene el siguiente obteniendo la recta tangente al punto $\left(x_{n},\,f(x_{n})\right)$ de la gráfica de $f$. Dicha recta por definición tiene pendiente $f'(x_{n})$, y como pasa por $x_{n}$, la ecuación de esa recta tangente es: 

$$y_{\mathrm{tangente},\,n}(x)=f(x_{n})+(x-x_{n})f'(x_{n})$$

El siguiente punto de la sucesión, $x_{n+1}$ es simplemente la intersección de la recta tangente $y_{\mathrm{tangente},\,n}(x)$ con el eje $x$ (i.e. $y_{\mathrm{tangente},\,n}(x_{n+1})=0$. Por ende, $x_{n+1}$ es:

$$x_{n+1}=x_{n}-\frac{f(x_{n})}{f'(x_{n})},$$

donde $n$ es un entero mayor o igual a 0. 

De ese modo, implementamos el método de Newton de forma recursiva con la siguiente función:

```julia
function newton(f::Function,deriv_f::Function,adivinanza::Real)
        tolerancia=0.0000000000001
        xn=adivinanza
        if deriv_f(xn)==0
            xn=xn+1
        end
        if abs(f(xn))<=tolerancia
            xn
        else
            xn=xn-(f(xn)/deriv_f(xn))
            newton(f,deriv_f,xn)
        end
    end
```

Esta función toma como argumentos a la función $f$ y su derivada $f'$ (que por ahora se implementa a mano en otra función). La función simplemente regresa el valor de $x_{n}$ si $|f(x_{n})|$ es menor al valor de tolerancia. Para evitar divisiones entre cero, la función verifica si $f'(x_{n})$ llega a ser cero (ya que eso daría una recta tangente horizontal que no intersecta con el eje $x$, haciendo imposible calcular $x_{n+1}$ en primer lugar). En ese caso, cambia el valor de $x_{n}$ a $x_{n}+1$, para pasar a un punto donde $f'$ no sea cero.

#### Inciso (c)

Ahora, buscamos las soluciones de la ecuación $\tan(x)=x$. Para ello, podríamos usar $f(x)=\tan(x)-x$ y usarla, junto a su derivada, como argumento de nuestra función newton. Sin embargo, eso puede causar problemas porque la función tangente diverge. Mejor, exploremos una condición alternativa para que $x$ cumpla la ecuación, usando la definición de tangente:

$$\tan(x)=x$$
$$\longleftrightarrow\tan(x)-x=0$$
$$\longleftrightarrow\frac{\sin(x)}{\cos(x)}-x=0$$

Luego, simplemente debido a que la ecuación original sólo tiene sentido la igualdad mientras $\cos(x)\neq0$, podemos dar el siguiente paso de equivalencia, que es multiplicar ambos lados de la última expresión por $\cos(x)$ para llegar finalmente a:

$$\sin(x)-x\cos(x)=0$$

O sea, en resumen, $\tan(x)=x\longleftrightarrow\sin(x)-x\cos(x)=0$.

Por ende usamos la función $f(x)=\sin(x)-x\cos(x)$ como la función f a la que debemos encontrarle raíces usando el método de Newton. Su derivada es $f'(x)=\cos(x)-\cos(x)+x\sin(x)=x\sin(x)$. De ese modo, estamos listos para implementar el método de Newton:

```julia
newton(x->sin(x)-x*cos(x),x->x*sin(x),1)
```

Verifiquemos gráficamente si coincide con alguna intersección entre $\sin(x)$ y $x\cos(x)$ (o bien, entre $x$ y $\tan(x)$).

```julia
figure(figsize=(5,4))
plot([-15:0.0001:15],[sin(x) for x in [-15:0.0001:15]],"g-",label=L"$\sin(x)$")
plot([-15:0.0001:15],[x*cos(x) for x in [-15:0.0001:15]],"r-",label=L"$x\cos(x)$")
legend()
```

```julia
figure(figsize=(5,4))
plot([-15:0.0001:15],[tan(x) for x in [-15:0.0001:15]],"g-",label=L"$\tan(x)$")
plot([-15:0.0001:15],[x for x in [-15:0.0001:15]],"r-",label=L"$x$")
axis([-15,15,-15,15])
legend()
```

Comprobamos que es más fácil encontrar las soluciones a la ecuación $\tan(x)=x$ usando la primera gráfica (que es lo mismo que encontrar las raíces de $f(x)=\sin(x)-x\cos(x)$) que usando la segunda (que es lo mismo que encontrar las raíces de $g(x)=\tan(x)-x$), ya que la función $\tan(x)$ diverge y además la función predeterminada tan() deja en la gráfica una línea vertical donde $\tan(x)$ diverge. 

Para finalizar, implementamos una función que nos dé las raíces de esta función (o de cualquier otra función $f$) en un intervalo $[a\,b]$ usando el método de Newton, que tome a la función $f$, a su derivada $f'$ y a los extremos del intervalo $a$ y $b$ como argumentos.

```julia
function raices_newton(f::Function,deriv_f::Function,A::Real,B::Real)
    tolerancia=0.00000000001
    N=20
    contador=0
    raiz1=0
    raiz2=1000000000000000
    if abs(A-B)<=tolerancia
        error("intervalo no válido")
    elseif A>B
        a=B
        b=A
    else
        a=A
        b=B
    end
    h=(b-a)/N
    for n in 0:(N-1)
        an=a+n*h
        bn=an+h
        cn=(an+bn)/2
        if abs(deriv_f(cn))>tolerancia
        raiz1=newton(f,deriv_f,cn)
        end        
        if abs(raiz1-raiz2)>tolerancia && an<raiz1<bn
            raiz2=raiz1
            contador=contador+1
            println(raiz2)
        end
    end
    if contador==0
        println("No se encontraron raices")
    end
end
```

Como la función biseccion, la función raices_newton construye el intervalo $[a,\,b]$ a partir de los valores reales que se introducen como sus dos últimos argumentos. Luego, parte el intervalo en $N$ subintervalos de igual longitud, $[a_{n},\,b_{n}]$. En cada uno, haya el punto medio $c_{n}$ y verifica que $|f'(c_{n})|$ sea mayor a valor de tolerancia. Si sí lo es lo introduce como el valor de la adivinanza en la función newton, y el resultado lo asigna como valor de una variable llamada raiz1 ($r_{1}$ en lenguaje matemático). Tenemos una segunda variable llamada raiz2 ($r_{2}$) cuyo valor se inicializa en uno muy diferente al de raiz1. La siguiente parte del algoritmo, que usa tanto a raiz1 como a raiz2, sirve para evitar imprimir raíces iguales más de una vez al terminar de ejecutar la función. Para ello, primero verifica que el último valor calculado de raiz1 (usando el método de Newton con la función Newton) esté en el intervalo $[a_{n},\,b_{n}]$ (esto se implementa porque generalmente si se usa $c_{n}$ como adivinanza, la raíz calculada con el método de Newton puede quedar fuera del intervalo, y nos interesa que esté dentro de él) y que además sea distinto al de raiz2. Si ambas condiciones se cumplen, raiz2 adopta el valor de raiz1, se incrementa el contador y se imprime a raiz2 (que es una raíz). De ese modo, si el método llegase a calcular una raíz que sea igual a una calculada previamente, simplemente no la cuenta y no la imprime. Al final, si el valor del contador no incremento, se regresa un mensaje que dice que no se encontraron raíces en el intervalo especificado.

Probamos nuestra función para ver las raíces en el intervalo [-15,15], que es el que graficamos más arriba.

```julia
raices_newton(x->sin(x)-x*cos(x),x->x*sin(x),-15,15)
```

Vemos que los resultados arrojados coinciden con las intersecciones entre las gráficas de las funciones $\sin(x)$ y $x\cos(x)$ (y también las intersecciones entre las de $\tan(x)$ y $x$, exceptuando las rectas verticales trazados por la función tan donde $\tan(x)$ diverge), a excepción de los valores $\pm6.455\ldots\times10^{-5}$, que en realidad deberían ser sólo un $0$ (sin embargo, el que aparecieran ellos y no cero es debido al valor de tolerancia que tiene la función newton). De ese modo, verificamos que esta función que sirve para encontrar las soluciones de la ecuación $\tan(x)=x$ en un intervalo dado sí funciona.

Finalmente, veamos qué ocurre si especificamos un intervalo donde no hay raíces, para ver si raices_newton regresa el mensaje adecuado.

```julia
raices_newton(x->sin(x)-x*cos(x),x->x*sin(x),2,3)
```

Y efectivamente, lo hace, por lo que raices_newton está bien implementada.
