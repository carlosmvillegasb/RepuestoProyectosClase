```julia
using PyPlot
```

```julia
fluctuaciones(t,θ,ω=1,r=1)=(1/2)*sqrt(cosh(2*r)-sinh(2*r)*cos(2*ω*t-θ))
```

```julia
real(1+im)
```

```julia
expectacion(t,θ,ω=1,r=1,α=1)=real(exp(im*ω*t)*(α'*cosh(r)-α*exp(-im*θ)*sinh(r))+exp(-im*ω*t)*(α*cosh(r)-exp(im*θ)*sinh(r)*α'))
```

```julia
expectacion(3,0,1,1,8+3im)
```

```julia
Qmax(t,θ,ω=1,r=1,α=1)=expectacion(t,θ,ω,r,α)+fluctuaciones(t,θ,ω,r)
```

```julia
Qmin(t,θ,ω=1,r=1,α=1)=expectacion(t,θ,ω,r,α)-fluctuaciones(t,θ,ω,r)
```

```julia
T=[0:0.01:10];
```

```julia
plot(T,[Qmax(t,0,1,1,1+im) for t in T],"b--")
plot(T,[Qmin(t,0,1,1,1+im) for t in T],"b--")
plot(T,[expectacion(t,0,1,1,1+im) for t in T],"r-")
xlabel(L"$t$")
ylabel(L"$\langle\hat{Q}(t)\rangle$")
title("Valor de expectación de la cuadratura respecto al tiempo con "L"$\theta=0$")
axis("tight")
```

```julia
plot(T,[Qmax(t,π,1,1,1+im) for t in T],"b--")
plot(T,[Qmin(t,π,1,1,1+im) for t in T],"b--")
plot(T,[expectacion(t,π,1,1,1+im) for t in T],"r-")
xlabel(L"$t$")
ylabel(L"$\langle\hat{Q}(t)\rangle$")
title("Valor de expectación de la cuadratura respecto al tiempo con "L"$\theta=\pi$")
axis("tight")
```

```julia
plot(T,[Qmax(t,0,1,1,1+im) for t in T],"b--")
plot(T,[Qmin(t,0,1,1,1+im) for t in T],"b--")
plot(T,[expectacion(t,0,1,1,1+im) for t in T],"r-",label=L"$\theta=0$")
plot(T,[Qmax(t,π,1,1,1+im) for t in T],"m--")
plot(T,[Qmin(t,π,1,1,1+im) for t in T],"m--")
plot(T,[expectacion(t,π,1,1,1+im) for t in T],"g-",label=L"$\theta=\pi$")
xlabel(L"$t$")
ylabel(L"$\langle\hat{Q}(t)\rangle$")
title("Valor de expectación de la cuadratura respecto al tiempo con diferentes valores de "L"$\theta$")
axis("tight")
legend(loc=0)
```
