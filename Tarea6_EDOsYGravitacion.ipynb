#### Inciso (i)

Definimos las funciones que den un solo paso para cada método:

```julia
function euler(f::Function, x0, t0::Real, h::Real)
    xn=x0+h*f(x0,t0)
    return xn
end
```

```julia
function puntomedio(f::Function, x0, t0::Real, h::Real)
    k1=f(x0,t0)
    k2=f(x0+(h/2)*k1,t0+(h/2))
    xn=x0+h*k2
    return xn
end
```

```julia
function RK4(f::Function, x0, t0::Real, h::Real)
    k1=f(x0,t0)
    k2=f(x0+(h/2)*k1,t0+(h/2))
    k3=f(x0+(h/2)*k2,t0+(h/2))
    k4=f(x0+h*k3,t0+h)
    xn=x0+(h/6)*(k1+2*k2+2*k3+k4)
    return xn
end
```

#### Inciso (ii)

Ahora, implementamos la función `integrar` que acepte como argumentos el método para integrar, la función a integrar, el rango de tiempos (dado por el tiempo inicial y el tiempo final) y el tamaño del paso, además del valor inicial de la variable independiente.

```julia
function integrar(Metodo::Function,f::Function, x0, t0::Real, tf::Real, h::Real)
    T=Float64[t0]
    X=Vector{Float64}[[x0]]
    xn=x0
    for ti in (t0+h):h:tf
        xn=Metodo(f,xn,ti-h,h)
        push!(T,ti)
        push!(X,[xn])
    end
    return T,hcat(X...)'
end
```

En ella, al tiempo t0 se llena la columna de tiempos, T, con t0 y la de la variable, X, con x0. Luego, se inicia el valor de `xn` en `x0`. Luego, se aplica el método para encontrar los valores de x a los tiempos ti mayores a t0 hasta llegar tf (tienen que ser MAYORES que t0 porque t0 ya se llenó desde el principio, por eso se usa el rango `(t0+h):h:tf`). Luego, el chiste es llenar T con ti y X con el valor de x que le corresponde a ti. Como la función método acepta el valor actual de la variable a un tiempo dado y regresa el nuevo, y como el tiempo ti es el tiempo nuevo, el tiempo actual debe ser ti, y el último valor asignado a `xn` es el valor actual de x. Por eso se usa `Metodo(f,x0,ti-h,h)`. Como el primer tiempo es t0 y el primer valor de xn es el dado como argumento, x0, el nuevo valor de `xn` es `Metodo(f,x0,t0,h)` y ese es el que corresponde a `t0+h` por definición de cómo se construyeron las funciones `Metodo` y coincide en cómo lo asigna la función `integrar`, ya que `t0+h` es grabado en T al mismo tiempo que `Metodo(f,x0,t0,h)` queda grabado en X. Así hasta todos los tiempos hasta llegar a `tf`. tf es el tiempo nuevo, mientras que tf-h es el tiempo actual y el último valor de xn es el valor actual. El nuevo valor (último valor) entonces debe ser `Metodo(f,x0,tf-h,h)` y es el que corresponde a tf. Al final, `tf` es grabado en T y `Metodo(f,x0,tf-h,h)` es grabado en X como el valor que le corresponde. 

#### Inciso (iii)

La ecuación diferencial del oscilador armónico simple es:
$$\frac{d^{2}x}{dt^{2}}=-\omega^2 x$$
Que puede pasarse a la forma $\dot{\vec{x}}=f(\vec{x},t)$ con:
$$\dot{x}_{1}=x_{2}$$
$$\dot{x}_{2}=-\omega^2 x_{1}$$

Definimos la función que representa a la ecuación diferencial del oscilador armónico simple, donde podamos variar $\omega$:

```julia
OsciladorArmonico(X,t,ω=1)=[X[2],-(ω^2)*X[1]]
```

```julia
function OsciladorArmonicoF(ω)
    funct(X,t)=OsciladorArmonico(X,t,ω)
    return funct
end
```

Así, `OsciladorArmonicoF(ω)` es ya una función que puede servir de argumento para los diferentes métodos para cualquier valor de `ω`.

Ahora, vemos los resultados que obtenemos al aplicar los diferentes métodos para un mismo valor del paso h, el mismo intervalo de tiempos, las mismas condiciones iniciales y el mismo valor de $\omega$.

```julia
T,X=integrar(euler,OsciladorArmonicoF(10),[3,0],0,15,0.001)
plot(T,X[:,1])
xlabel(L"$t\,\mathrm{(s)}$")
ylabel(L"$x(t)\,\mathrm{(cm)}$")
title("Solución del oscilador armónico usando método de Euler")
```

```julia
T,X=integrar(puntomedio,OsciladorArmonicoF(10),[3,0],0,15,0.001)
plot(T,X[:,1])
xlabel(L"$t\,\mathrm{(s)}$")
ylabel(L"$x(t)\,\mathrm{(cm)}$")
title("Solución del oscilador armónico usando método de integración del punto medio")
```

```julia
maximum(X[:,1])
```

```julia
T,X=integrar(RK4,OsciladorArmonicoF(10),[3,0],0,15,0.001)
plot(T,X[:,1])
xlabel(L"$t\,\mathrm{(s)}$")
ylabel(L"$x(t)\,\mathrm{(cm)}$")
title("Solución del oscilador armónico usando método RK4")
```

Puede observarse que la solución de los tres métodos es distinta. El método de Euler no conserva la energía y muestra que la amplitud del oscilador armónico simple aumenta con el tiempo, cosa que es cualitativamente incorrecta. El método del punto medio sí parece conservar la energía, pero da un valor de la amplitud ligeramente mayor que el valor de la amplitud que debería obtenerse de la condición inicial. Mientras tanto, el método RK4 sí respeta ese valor de la amplitud y lo mantiene constante.

### Respuesta 2

Las ecuaciones de movimiento que deben resolverse son:
$$\ddot{\vec{r}}_{1}=\frac{Gm_{2}}{|\vec{r}_{1}-\vec{r}_{2}|^{3}}(\vec{r}_{2}-\vec{r}_{1})$$
$$\ddot{\vec{r}}_{2}=\frac{Gm_{1}}{|\vec{r}_{1}-\vec{r}_{2}|^{3}}(\vec{r}_{1}-\vec{r}_{2})$$

Que al pasarse a un sistema de la forma $\dot{\vec{Q}}=f(\vec{Q},t)$, con $Q=(x_{1},\,y_{1},\,\dot{x}_{1},\,\dot{y}_{1},x_{2},\,y_{2},\,\dot{x}_{2},\,\dot{y}_{2})$, queda:

$$\dot{Q}_{1}=Q_{3}$$
$$\dot{Q}_{2}=Q_{4}$$
$$\dot{Q}_{3}=Gm_{2}(Q_{5}-Q_{1})/\left((Q_{1}-Q_{5})^{2}+(Q_{2}-Q_{6})^{2}\right)^{3/2}$$
$$\dot{Q}_{4}=Gm_{2}(Q_{6}-Q_{2})/\left((Q_{1}-Q_{5})^{2}+(Q_{2}-Q_{6})^{2}\right)^{3/2}$$
$$\dot{Q}_{5}=Q_{7}$$
$$\dot{Q}_{6}=Q_{8}$$
$$\dot{Q}_{7}=Gm_{1}(Q_{1}-Q_{5})/\left((Q_{1}-Q_{5})^{2}+(Q_{2}-Q_{6})^{2}\right)^{3/2}$$
$$\dot{Q}_{8}=Gm_{1}(Q_{2}-Q_{6})/\left((Q_{1}-Q_{5})^{2}+(Q_{2}-Q_{6})^{2}\right)^{3/2}$$


La implementamos como:

```julia
function DosCuerpos(Q,t,m1=1,m2=1)
    G=0.0001173
    d=sqrt(((Q[1]-Q[5])^2)+((Q[2]-Q[6])^2))
    D=d^3
    V=float64(big([Q[3],Q[4],G*m2*(Q[5]-Q[1])/D,G*m2*(Q[6]-Q[2])/D,Q[7],Q[8],G*m1*(Q[1]-Q[5])/D,G*m1*(Q[2]-Q[6])/D]))
    return V
end
```

```julia
function DosCuerposM(m1,m2)
        funcm(X,t)=DosCuerpos(X,t,m1,m2)
    return funcm
end
```

De ese modo `DosCuerposM(m1,m2)` es una función válida como argumento de `integrar` una vez elegidos los valores de `m1` y `m2`.

Nótese que se le dio un valor a la constante $G$ distinto al que tiene en SI. Eso es porque se decidió poner la constante $G$ en unidades más convenientes. La unidad de masa en este problema será la masa de la tierra ($M_{T}=5.97\times10^{24}\,\mathrm{kg}$), la unidad de distancia es la unidad astronómica ($1\,\mathrm{uA}=1.5\times10^{11}\,\mathrm{m}$), y la unidad de tiempo es un año ($1\,\mathrm{yr}=3.1536\times10^{7}\,\mathrm{s}$). De ese modo $G=1.173\times10^{-4}\,\mathrm{uA^{3}M_{T}^{-1}yr^{-2}}$.

Probamos resolver para el sistema Sol-Tierra, usando datos de la velocidad orbital terrestre (que se usa como condición inicial) y la masa del Sol:

```julia
T,Q=integrar(RK4,DosCuerposM(1,332950),[1,0,0,6.25,0,0,0,0],0,1,0.001)
X1=Q[:,1]
Y1=Q[:,2]
X2=Q[:,5]
Y2=Q[:,6]
plot(X1,Y1,"r-")
plot(X2,Y2,"b.")
axis("equal")
xlabel(L"$x\,\mathrm{(uA)}$")
ylabel(L"$y\,\mathrm{(uA)}$")
```

Probemos ahora para otras condiciones iniciales y masas:

```julia
T,Q=integrar(RK4,DosCuerposM(2,3),[0.0116,0,0,0.041,-0.0014,0,0,-0.049],0,3,0.0001)
X1=Q[:,1]
Y1=Q[:,2]
X2=Q[:,5]
Y2=Q[:,6]
plot(X1,Y1,"r-")
plot(X2,Y2,"b-")
axis("equal")
xlabel(L"$x\,\mathrm{(uA)}$")
ylabel(L"$y\,\mathrm{(uA)}$")
```

Vemos que es un comportamiento bastante extraño, pero eso es porque el centro de masas del sistema está desplazándose dentro de nuestro sistema de referencia, y al mismo tiempo, los planetas están ambos orbitando alrededor de él.

Si queremos una animación más familiar (en donde el centro de masas esté en el origen), deben ponerse ciertas restricciones a las condiciones iniciales. En este caso, son simplemente: 
$$m_{1}x_{1}+m_{2}x_{2}=0$$
$$m_{1}y_{1}+m_{2}y_{2}=0$$

La segunda ecuación puede cumplirse simplemente dando los valores iniciales $y_{1}=y_{2}=0$.

```julia
T,Q=integrar(RK4,DosCuerposM(2,3),[0.0116,0,0,0.041,-0.0077,0,0,-0.049],0,3,0.0001)
X1=Q[:,1]
Y1=Q[:,2]
X2=Q[:,5]
Y2=Q[:,6]
plot(X1,Y1,"r-")
plot(X2,Y2,"b-")
axis("equal")
xlabel(L"$x\,\mathrm{(uA)}$")
ylabel(L"$y\,\mathrm{(uA)}$")
```

Vemos que esa condición no fue suficiente, ya que era sólo para que la posición inicial del centro de masas esté en el origen. También debe ponerse una restricción en las velocidades iniciales, para que la velocidad del centro de masa sea cero en todo momento, y se quede fijo en el origen. Es simplemente cambiar las condiciones anteriores reemplazando posiciones iniciales por velocidades iniciales.

$$m_{1}u_{1}+m_{2}u_{2}=0$$
$$m_{1}v_{1}+m_{2}v_{2}=0$$
donde $u$ son las componentes horizontales de la velocidad y $v$ es la vertical. En todos los ejemplos hemos hecho $u_{1}=u_{2}=0$.

Vemos que muy rápidamente se obtiene una órbita hiperbólica y los planetas se alejan rápidamente uno del otro.

Podemos tener un mayor control sobre la forma de las órbitas de los planetas, vistas desde el sistema de referencia del centro de masa. Lo único que necesitamos es tomar como parámetros las masas de los planetas, y la excentricidad $\epsilon$ y semilado recto $l$ de la órbita de uno de ellos, e incluso podemos calcular el periodo $T$ de la orbita de los planetas del sistema, para poder hacer animaciones. Para ello, debemos tomar en cuenta lo siguiente:

En el marco de referencia del centro de masa, la línea que une a ambos planetas gira con cierta velocidad angular en cada momento, y por definición, ésta siempre pasa por el centro de masa. De ese modo, ambos planetas tienen en todo momento la misma velocidad angular respecto al centro de masa. El valor inicial de dicha velocidad angular se obtiene igualando la fuerza de atracción con la fuerza centrípeta, de modo que los valores iniciales de las velocidades de cada planeta son:
$$v_{1}=\frac{\sqrt{Gm_{2}|x_{1}|}}{|x_{1}-x_{2}|},\qquad v_{2}=-\frac{\sqrt{Gm_{1}|x_{2}|}}{|x_{1}-x_{2}|}$$
donde $x_{1}$ y $x_{2}$ son las posiciones iniciales de los planetas, asumiendo que parten del eje $x$. Además, deben cumplir $m_{1}x_{1}+m_{2}x_{2}=0$. Sin embargo, esas velocidades son las velocidades iniciales de órbitas circulares (donde la fuerza centrípeta iguala a la gravitacional). Para obtener formas diferentes, y al mismo tiempo mantener fijo al centro de masas, multipliquemos ambas velocidades por un factor $f$:
$$v_{1}=f\frac{\sqrt{Gm_{2}|x_{1}|}}{|x_{1}-x_{2}|},\qquad v_{2}=-f\frac{\sqrt{Gm_{1}|x_{2}|}}{|x_{1}-x_{2}|}$$

Como ya tenemos los valores iniciales de las posiciones y velocidades de los planetas, podemos conocer la magnitud del momento angular $L$ y la energía mecánica del sistema $E$, que son ambas constantes, a partir de esos valores iniciales y el factor $f$. Puede demostrarse que (y de hecho lo hice en sucio en mi cuaderno, usando solo las ecuaciones anteriores y la condición de dejar fijo el centro de masa en el origen):

$$L=f\sqrt{Gm_{1}^{2}m_{2}|x_{1}|}$$
$$E=\frac{Gm_{1}m_{2}}{x_{1}(1+\frac{m_{1}}{m_{2}})}\left(\frac{f^{2}}{2}-1\right)$$

Usando las expresiones del semilado recto $l$ y la excentricidad $\epsilon$, dadas por:
$$l=\frac{L^{2}(m_{1}+m_{2})}{Gm_{1}^{2}m_{2}^{2}}$$
$$\epsilon=\sqrt{1+\frac{2lE}{Gm_{1}m_{2}}}$$

Obtenemos ambos en términos de las condiciones iniciales y las masas del sistema, y puede demostrarse que quedan dados por las siguientes relaciones:
$$f=\sqrt{1+\epsilon}$$
$$x_{1}=\frac{l}{\left(1+\frac{m_{1}}{m_{2}}\right)f^{2}}$$

Finalmente, con el mismo valor del semilado recto $l$ y la excentricidad $\epsilon$ puede obtenerse el valor del semieje mayor $a$ con la relación geométrica $a=\frac{l}{|\epsilon^{2}-1|}$ si $\epsilon\neq1$ y $a=\frac{l}{2}$ si $\epsilon=1$. Una vez obtenido $a$, el periodo de las órbitas puede obtenerse simplemente con tercera ley de Kepler. 

Todos los resultados y condiciones podemos juntarlos en la siguiente función que toma los parámetros de las órbitas (las masas de los planetas y el semilado recto y excentricidad de la órbita) y devuelve las condiciones iniciales en las que debe evaluarse la función `integrar`, así como el tamaño $T$ del intervalo donde se integrará para poder observar un periodo del sistema:

```julia
function VelocidadesComp(m1,m2,l,ex)
    f=sqrt(1+abs(ex))
    G=0.0001173
    r1=l/((1+(m1/m2))*f^2)
    r2=-(m1/m2)*r1
    v1=f*sign(r1)*sqrt(G*m2*abs(r1))/abs(r1-r2)
    v2=f*sign(r2)*sqrt(G*m1*abs(r2))/abs(r1-r2)
    if ex==1
        a=l/2
    else
        a=l/abs(1-(ex^2))
    end
    T=2*pi*sqrt((a^3)/(G*(m1+m2)))
    return r1,r2,v1,v2,T
end
```

Probemos que esta las condiciones iniciales dadas por esta función evaluadas en la función `integrar` efectivamente nos devuelven las órbitas dadas por los diferentes parámetros:

```julia
VelocidadesComp(1,1,1,0.9)
```

```julia
m1,m2,l,ex=1,1,1,0.9
r1,r2,v1,v2,t=VelocidadesComp(m1,m2,l,ex)
T,Q=integrar(RK4,DosCuerposM(m1,m2),[r1,0,0,v1,r2,0,0,v2],0,t,t/1000)
X1=Q[:,1]
Y1=Q[:,2]
X2=Q[:,5]
Y2=Q[:,6]
plot(X1,Y1,"r-")
plot(X2,Y2,"b-")
axis("equal")
xlabel(L"$x\,\mathrm{(uA)}$")
ylabel(L"$y\,\mathrm{(uA)}$")
```

```julia
using ODE
```

`ODE.jl` funciona igual que los métodos que hemos implementado. Las funciones de ODE aceptan una función $F(t,x)$ que describe a la ODE, el valor inicial de las variables y el intervalo de tiempos.

Definimos la función `RLC` que describe al circuito RLC, cuya ecuación está dada por:

$$\frac{d^{2}q}{dt^{2}}+\frac{R}{L}\frac{dq}{dt}+\frac{1}{LC}q=\frac{V_{0}}{L}\sin(\omega_{0}t)$$

Definimos la función que representa a este sistema:

```julia
CircuitoRLC(t,X,R=1,L=1,C=1,V0=1,ω0=1)=[X[2],-(R/L)*X[2]-(1/(L*C))*X[1]-(V0/L)*sin(ω0*t)]
```

```julia
function CircuitoRLCParametros(R,L,C,V0,ω0)
    funcR(t,X)=CircuitoRLC(t,X,R,L,C,V0,ω0)
    return funcR
end
```

Resolvemos usando las funciones del paquete, para encontrar la corriente en función del tiempo. Como condición pondremos que el capacitor está descargado y la corriente es distinta de cero.  

```julia
Tiempo, Corr=ode45(CircuitoRLCParametros(230,0.38,0.0000001,120,60),[0.0,3.0],(0,0.02))
Corrientes=[C[2] for C in Corr]
plot(Tiempo,Corrientes)
xlabel(L"$t\,\mathrm{(s)}$")
ylabel(L"$I\,\mathrm{(A)}$")
axis("tight")
```

De ese modo, se pudo resolver la ecuación del circuito RLC en serie usando parámetros reales.
