```julia
function euler(y0,t0,h,f,N)
    y = Vector{Float64}[[y0]]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t,y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,[y0])
    end
    t,hcat(y...)'
end
```

```julia
tiroparabolico_friccioncuadratica(t,X,γ=0.02)=[X[2],-γ*X[2].*sqrt((X[2].^2)+(X[4].^2)),X[4],-9.81-γ*X[4].*sqrt((X[2].^2)+(X[4].^2))]
```

```julia
function tiroparabolico_friccioncuadraticaG(γ)
    funct2(t,X)=tiroparabolico_friccioncuadratica(t,X,γ)
    return funct2
end
```

```julia
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v)
    q = Vector{Float64}[[q0]]
    v = Vector{Float64}[[v0]]
    t = Float64[t0]
    for ti in t0:h:tf
        v_nueva = v0 + h*f_v(ti,q0,v0)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
    end
    t,hcat(q...)',hcat(v...)'
end
```

```julia
function parafriccromer_q(t,q,v)
    [v[1],v[2]]
end
```

```julia
function parafriccromer_v(t,q,v)
    g=9.81
    γ=0.03
    [-γ*v[1]*sqrt((v[1]^2)+(v[2]^2)),-g-γ*v[2]*sqrt((v[1]^2)+(v[2]^2))]
end
```

```julia
t,Q,V=euler_cromer([0,0],[30,70],0,15,0.001,parafriccromer_q,parafriccromer_v)
plot(Q[:,1],Q[:,2])
```

```julia
function pendcromer_q(t,q,v)
    [v[1]]
end
```

```julia
function pendcromer_v(t,q,v)
    g=9.81
    L=13.5
    [-g/L*sin(q[1])]
end
```

```julia
t,Q,V=euler_cromer([pi-0.01],[0],0,15,0.001,pendcromer_q,pendcromer_v)
plot(Q[:,1],V[:,1])
```
