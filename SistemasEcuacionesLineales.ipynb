$\mathbb{A}$ es una matriz de $m$ renglones y $n$ columnas con $m=n$. $\mathbb{x}$ debe ser un vector columna de $n$ renglones.

$$b_{i}=\sum_{k=1}^{n} A_{ik}x_{k}$$

Así si $\mathbb{U}$ es triangular superior ($U_{ij}=0$ si $i>j$), entonces: 

$$c_{i}=\sum_{j=i}^{n} U_{ij}x_{j}$$

La solución es:

$$x_{n}=\frac{1}{U_{nn}}C_{n}$$
$$x_{i}=\frac{1}{U_{ii}}\left(C_{i}-\sum_{j=i+1}^{n}U_{ij}x_{j}\right)\qquad n> i \geq1$$

```julia
function ResuelveTriS(U,C)
if size(C,2)!=1
    error("No es vector columna")
end
if size(U,1)!=size(U,2)
    error("Matriz no cuadrada")
end
if size(U,1)!=length(C)
    error("Matriz y vector de tamaños distintos")
end
n=size(U,1)
X=zeros(n,1)
X[n]=(1/U[n,n])*C[n]
    for k in 2:n
        Prodj=[U[n-k+1,j]*X[j] for j in (n-k+2):n]
        X[n-k+1]=(1/U[n-k+1,n-k+1])*(C[n-k+1]-sum(Prodj))
    end
    return X
end
```

Ahora, hagamos un método que nos permita reducir cualquier matriz a una mtriz triangular superior, de modo que podamos usar la función anterior para resolver el sistema de ecuaciones en cuestión.

```julia
function Triangulariza(A,b)
    A=float(A)
    b=float(b)
    m=size(A,1)
    n=size(A,2)
    if m!=n
        error("MAtriz no cuadrada")
    end
    if size(b,2)!=1
    error("No es vector columna")
    end
    if size(A,1)!=length(b)
    error("Matriz y vector de tamaños distintos")
    end
    for i in 1:(n-1)
        for k in 1:(n-i)
            coef=(float(A[i+k,i])/float(A[i,i]))
            b[i+k,1]=float(b[i+k,1])-float(b[i,1]).*coef
            for j in 1:n
                A[i+k,j]=float(A[i+k,j])-float(A[i,j]).*coef
            end
        end       
    end
   # println(round(A,2))
    #println(round(b,2))
    return A,b
end
```

```julia
function ResuelveEcuacionesLineales(A,b)
    U,C=Triangulariza(A,b)
    X=ResuelveTriS(U,C)
    return X[:,1]
end
```

```julia
@time ResuelveEcuacionesLineales([3 7; 2 4],[1,1])
```

```julia
@time ResuelveEcuacionesLineales([2 1 -1; -3 -1 2; -2 1 2],[8,-11,-3])
```

### Ecuacion de Laplace

El sistema de ecuaciones es:

$$V_{1}=V(a)$$
$$\frac{1}{2}V_{i-1}-V_{i}+\frac{1}{2}V_{i+1}=0;\qquad 1<i<N+1$$
$$V_{N+1}=V(b)$$

Donde $N$ es el número de subintervalos en que se dividió el intervalo el intervalo $[a,\,b]$, con $$x_{i}=a+(i-1)\frac{b-a}{N};\quad 1\leq i\leq N+1$$ y $V(x_{i})=V_{i}$

```julia
function LaplaceCapacitor(a::Real,b::Real,Va::Real,Vb::Real,h::Real)
    X=[a:h:b]
    n=length(X)
    b=zeros(n)
    A=zeros(n,n)
    b[1]=Va
    b[n]=Vb
    A[1,1]=1
    A[n,n]=1
    for i in 2:(n-1)
        A[i,i-1]=1/2
        A[i,i]=-1
        A[i,i+1]=1/2
    end
    V=ResuelveEcuacionesLineales(A,b)
    return X,V
end
```

```julia
@time X,V=LaplaceCapacitor(5,10,50,100,0.01)
plot(X,V)
```

```julia
function PoissonPuntual(a::Real,b::Real,q::Real,h::Real)
    X=[a:h:b]
    n=length(X)
    b=zeros(n)
    A=zeros(n,n)
    c=int(ceil(n/2))
    b[c]=q*(h^2)/2
    A[1,1]=1
    A[n,n]=1
    for i in 2:(n-1)
        A[i,i-1]=1/2
        A[i,i]=-1
        A[i,i+1]=1/2
    end
    V=ResuelveEcuacionesLineales(A,b)
    return X,V
end
```

```julia
X,V=PoissonPuntual(5,10,5,0.01)
plot(X,V)
```

### Ecuación de Poisson

El laplaciano en diferencias finitas es:

$$\nabla^{2}f=\frac{f(x_{0}+h,\,y_{0})+f(x_{0}-h,\,y_{0})+f(x_{0},\,y_{0}+h)+f(x_{0},\,y_{0}-h)-4f(x_{0},\,y_{0})}{h^{2}}$$

```julia
function ConvertirIJaK(rengloni::Int,columnaj::Int,renglones::Int,columnas::Int)
    k=int(columnas*(rengloni-1)+columnaj)
    return k
end
```

```julia
function ConvertirKaIJ(elementok::Int,renglones::Int,columnas::Int)
    i=int(ceil(elementok/columnas))
    if elementok%columnas==0
        j=columnas
    else
        j=elementok%columnas
    end
    return i,j
end
```

Usamos
$$(x_{0}+h,\,y_{0})\rightarrow (i,\,j+1)$$
$$(x_{0}-h,\,y_{0})\rightarrow (i,\,j-1)$$
$$(x_{0},\,y_{0}+h)\rightarrow (i-1,\,j)$$
$$(x_{0},\,y_{0}-h)\rightarrow (i+1,\,j)$$

ya que subir $y$ en la malla implica bajar en el número del renglon de la matriz que representa a la malla.

$$\nabla^{2}V=\frac{V_{i,\,j+1}+V_{i,\,j-1}+V_{i-1,\,j}+V_{i+1,\,j}-4V_{i,\,j}}{h^{2}}$$

La malla tiene valores de $x$ que están en un arreglo `[xmin:h:xmax]` con $N$ elementos, y valores de $y$ que están en un arreglo `[ymin:h:ymax]` con $M$ elementos, de modo que la malla tiene $M$ renglones y $N$ columnas.

Usamos de ese modo la ecuación de Poisson para los puntos interiores de la malla, igualando $\nabla^{2}V$ a $f(x,\,y)\rightarrow f_{i,\,j}$:
$$V_{i,\,j+1}+V_{i,\,j-1}+V_{i-1,\,j}+V_{i+1,\,j}-4V_{i,\,j}=f_{i,\,j}h^{2}$$

Esa ecuación se traduce a una de las $K$ ecuaciones que se obtienen al convertir $(i,\,j)\rightarrow k$

$$P_{k(i,\,j+1)}+P_{k(i,\,j-1)}+P_{k(i-1,\,j)}+P_{k(i+1,\,j)}-4P_{k(i,\,j)}=F_{k(i,\,j)}h^{2};\quad 1<i<M,\,1<j<N$$

La ecuación anterior sólo aplica a los puntos interiores. Para los puntos de frontera, tenemos:
$$P_{k(i,\,1)}=0\quad 1\leq i\leq M$$
$$P_{k(i,\,N)}=0\quad 1\leq i\leq M$$
$$P_{k(1,\,j)}=0\quad 1\leq j\leq N$$
$$P_{k(M,\,j)}=0\quad 1\leq j\leq N$$

De ese modo, todas las ecuaciones anteriores son el sistema de ecuaciones que debe resolver la computadora para resolver la ecuación de Poisson. Podemos hacer una función que se encargue de implementar las condiciones de frontera en los bordes de la malla y de construir el sistema de ecuaciones que sale de éstas y de las ecuaciones para los puntos interiores.

```julia
function PoissonPuntual(xmin::Real,xmax::Real,ymin::Real,ymax::Real,h::Real,q::Real)
    #Construir la malla
    ax=min(xmin,xmax)
    bx=max(xmin,xmax)
    ay=min(ymin,ymax)
    by=max(ymin,ymax)
    h=abs(h)
    X=[ax:h:bx]
    Y=[ay:h:by]
    N=length(X)
    M=length(Y)
    #Construir las matrices del sistema de ecuaciones
    A=zeros(M,N)
    P=zeros(M*N,M*N)
    F=zeros(M*N)
    #Implementar las ecuaciones de la frontera en los bordes de la "caja"
    for i in 1:M
        F[ConvertirIJaK(i,1,M,N)]=0
        F[ConvertirIJaK(i,N,M,N)]=0
        P[ConvertirIJaK(i,1,M,N),ConvertirIJaK(i,1,M,N)]=1
        P[ConvertirIJaK(i,N,M,N),ConvertirIJaK(i,N,M,N)]=1
    end
    for j in 1:N
        F[ConvertirIJaK(1,j,M,N)]=0
        F[ConvertirIJaK(M,j,M,N)]=0
        P[ConvertirIJaK(1,j,M,N),ConvertirIJaK(1,j,M,N)]=1
        P[ConvertirIJaK(M,j,M,N),ConvertirIJaK(M,j,M,N)]=1
    end
    #Implementar las ecuaciones de los puntos interiores, poniendo los coeficientes
    for i in 2:(M-1)
        for j in 2:(N-1)
            P[ConvertirIJaK(i,j,M,N),ConvertirIJaK(i,j+1,M,N)]=1
            P[ConvertirIJaK(i,j,M,N),ConvertirIJaK(i,j-1,M,N)]=1
            P[ConvertirIJaK(i,j,M,N),ConvertirIJaK(i-1,j,M,N)]=1
            P[ConvertirIJaK(i,j,M,N),ConvertirIJaK(i+1,j,M,N)]=1
            P[ConvertirIJaK(i,j,M,N),ConvertirIJaK(i,j,M,N)]=-4       
            #F[ConvertirIJaK(i,j)]=h*h*distribucioncarga(i,j,M,N) seria lo que usariamos si para hacer que esta funcion resuelva el campo de cualquier distribucion de carga
        end
    end
    #Poner la carga puntual en el centro; esto no es necesario si usamos una funcion `distribucioncarga` que haya hecho lo mismo en el loop anterior
    cx=int(ceil(N/2))
    cy=int(ceil(M/2))
    F[ConvertirIJaK(cy,cx,M,N)]=-4*pi*q
    #Resolver para encontrar el potencial en todo punto
    Vk=ResuelveEcuacionesLineales(P,F)
    #Reconstruir la malla rectangular, convirtiendo el vector columna V_k a la matriz V_i,j
    for k in 1:length(Vk)
        i,j=ConvertirKaIJ(k,M,N)
        A[i,j]=Vk[k]
    end
    return A
end
```

Luego podremos explorar cómo implementar funciones que resuelvan la ecuación de Poisson para otras distribuciones de carga y para otras condiciones de frontera. Implementar otras distribuciones de carga es sencillo, ya que es sólo cómo llenar los valores $F_{k(i,\,j)}$ en la matriz $F$. Sin embargo, implementar otras condiciones de frontera es más complicado porque cambia totalmente el algoritmo de cómo llenar los coeficientes de las matrices del sistema de ecuaaciones lineales.
