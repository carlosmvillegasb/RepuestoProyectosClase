#### Inciso (a)

Volvemos a usar la función newton de la tarea pasada, pero ahora haciéndola iterativa en vez de recursiva (usando un ciclo while que se detiene cuando se logra que $|f(x_{n})|$ sea menor que la tolerancia o cuando se alcanza cierto número de iteraciones).

```julia
function newton(f::Function,deriv_f::Function,adivinanza)
        tolerancia=0.0000000000001
        xn=adivinanza
    contador=0
        if deriv_f(xn)==0
            xn=xn+1
        end
        if abs(f(xn))<=tolerancia
            xn
        else
        while abs(f(xn))>tolerancia || contador<300
            xn=xn-(f(xn)/deriv_f(xn))
            contador=contador+1
        end
        end
    return xn
    end
```

La volvemos a probar rápidamente para las funciones de la tarea pasada para ver que funciona bien:

```julia
newton(x->x^2-2,x->2x,5)
```

```julia
newton(x->sin(x)-x*cos(x),x->x*sin(x),4.5)
```

$$z_{3}=e^{i4\pi/3}=-\frac{1}{2}-\frac{\sqrt{3}}{2}i=-0.5-0.866025i$$

Tal parece que la función nueva ha convergido a alguna raíz con varios valores de la adivinanza. Veamos qué ocurre al usar la función recursiva de la tarea pasada.

```julia
function newton_viejo(f::Function,deriv_f::Function,adivinanza)
        tolerancia=0.0000000000001
        xn=adivinanza
        if deriv_f(xn)==0
            xn=xn+1
        end
        if abs(f(xn))<=tolerancia
            xn
        else
            xn=xn-(f(xn)/deriv_f(xn))
            newton(f,deriv_f,xn)
        end
    end
```

### Respuesta [2]

Tomamos en cuenta que podemos resolver cualquier sistema de $n$ ecuaciones de orden $m$ reduciéndolo a un sitema de $mn$ ecuaciones de primer orden:

$$\dot{\mathbf{x}}(t) = \mathbf{f}(\mathbf{x}(t), t),$$

donde $\mathbf{x} = (x_1, \ldots, x_n) \in \mathbb{R}^n$ es un vector y $\mathbf{f}: \mathbb{R}^{n+1} \to \mathbb{R}^n$ es una función vectorial.

En el caso del tiro parabólico, tenemos un sistema de dos ecuaciones (en dos variables) de segundo orden:

$$\ddot{x}=0$$
$$\ddot{y}=-g,$$

donde $g=9.81\,\mathrm{m/s^{2}}$ es la aceleración de la gravedad. 

Podemos reducirlo al siguiente sistema de ecuaciones, para poderlo resolver con el método de Euler general. Primero, como el del tiro parabólico es un sistema de 2 ecuaciones diferenciales de orden 2, necesitamos 4 variables en el sistema equivalente. Dichas variables las definimos como sigue:

$$x_{1}=x$$
$$x_{2}=\dot{x}$$
$$x_{3}=y$$
$$x_{4}=\dot{y}$$

De ese modo, el sistema de ecuaciones que describe al tiro parabólico es el siguiente:

$$\dot{x_{1}}=x_{2}$$
$$\dot{x_{2}}=0$$
$$\dot{x_{3}}=x_{4}$$
$$\dot{x_{4}}=-g$$

Esto ya tiene la forma $\dot{\mathbf{x}}(t) = \mathbf{f}(\mathbf{x}(t), t)$, donde $\mathbf{x}(t)=(x_{1}(t),\,x_{2}(t),\,x_{3}(t),\,x_{4}(t))$ y $\mathbf{f}(\mathbf{x}(t), t)=(x_{2}(t),\,0,x_{4}(t),-g)$.

#### Inciso (a)

Para construir la función que implementa el método general de Euler, necesitamos realizar funciones que tomen arreglos o matrices como argumento, ya que al vector $\mathbf{x}(t)$ hemos de representarlo con un areglo. Por el momento, necesitamos unas funciones que extraigan cierto renglón o cierta columna de una matriz  y lo regresen como un Array.

```julia
columnan(A,n)=A[1:size(A,1),n]
```

```julia
renglonm(A,m)=[x for x in A[m,1:size(A,2)]]
```

Finalmente, usamos estas funciones en la función `euler_general`, que toma una función vectorial como primer argumento (o sea, una función $\mathbf{F}$ que toma al arreglo que describe a $\mathbf{x}(t)$ y al escalar $t$ como argumentos y regresa un arreglo del mismo tamaño que el de $\mathbf{x}(t)$), como segundo argumento a un arreglo que contiene las condiciones iniciales de las variables del sistema de ecuaciones (y que son argumentos de $\mathbf{F}$), y dos argumentos reales para el tiempo inicial y el tiempo final. La función `euler_general` determina el número de variables `num_var` a partir del número de elementos en el arreglo de condiciones iniciales. Luego, crea un arreglo de tiempos desde el tiempo inicial hasta el tiempo final dando pasos de tamaño $h$, tras lo cual cuenta el tamaño de dicho arreglo y lo guarda en la variable `num_tiempos`. Luego, crea un arreglo para las posiciones,  `Xs`, con un número de renglones igual a  `num_var` y de columnas igual a `num_tiempos`. Acto seguido, empieza a aplicar el método de Euler. Primero llena la primera columna de `Xs` con los valores del arreglo de condiciones iniciales, pues la primera columna corresponde a $t$ igual al tiempo inicial. Posteriormente, a cada tiempo, va actualizando cada una de las varaibles del sistema de ecuaciones con el método de Euler. Al terminar, regresa el nuevo arreglo de tiempos y un arreglo cuyos elementos son los renglones de `Xs` (o sea, otro arreglo) que corresponden a los diferentes valores en el tiempo de cada una de las variables del sistema. Regresa estos dos arreglos como una tupla.

```julia
function euler_general(funcion_vectorial, arreglo_condiciones,tiempo_inicial,tiempo_final)
num_var=length(arreglo_condiciones)
    h=0.001
    tiempos=[tiempo_inicial:h:tiempo_final]
    num_tiempos=length(tiempos)
    Xs=zeros(num_var,num_tiempos)
    for m in 1:num_var
        Xs[m,1]=arreglo_condiciones[m]
    end
    for n in 2:num_tiempos
        for m in 1:num_var
            Xs[m,n]=Xs[m,n-1]+h*funcion_vectorial(columnan(Xs,n-1),tiempos[n-1])[m]
        end
    end
    return tiempos, [renglonm(Xs,m) for m in 1:num_var]
end
```

De ese modo, ya  estamos listos para implementar para implementar el método de Euler general para resolver numéricamente las ecuaciones del tiro parabólico y graficarlo.

La solución analítica del tiro parabólico es:

$$x(t)=x_{0}+ut$$
$$y(t)=y_{0}+vt-\frac{gt^{2}}{2}$$

Donde $x_{0}$, $u$, $y_{0}$ y $v$ son los valores iniciales de la posición horizontal, la velocidad horizontal, la posición vertical y la velocidad vertical, respectivamente. 

Para graficarla y comparar, simplemente copiamos e implementamos el código que se encargaba de calcular esas trayectorias usando la expresión analítica anterior.

```julia
const g=9.81
```

```julia
x(x0,u,t)=x0+u*t
```

```julia
y(y0,v,t)=y0+v*t-0.5*g*t^2
```

```julia
function tiro_parabolico(x0,u,y0,v,delta_t,t_final)
    ts=[0:delta_t:t_final]
    xs=[x(x0,u,t) for t in ts]
    ys=[y(x0,v,t) for t in ts]
    return ts,xs,ys
end
```

La funcion anterior tiene como valor fijo de tiempo inicial a $t=0$, lo cual coincide con el valor usado en la gráfica anterior.
```julia
ts,xs,ys=tiro_parabolico(20,30,30,70,0.001,15);
plot(xs,ys)
xlabel(L"$x(t)\,\mathrm{(m)}$")
ylabel(L"$y(t)\,\mathrm{(m)}$")
hlines(0, 0, 500, colors="k", linestyles="dashed")
vlines(0, -100, 300, colors="k", linestyles="dashed")
title("Tiro parabólico analítico")
```

Vemos que ambas gráficas son iguales, por lo que nuestro método numérico sí coincidió con la solución analítica.

#### Inciso (b)

Si ahora tenemos fricción proporcional a la velocidad, o sea $\mathrm{\vec{F}_{fricción}}=-b\mathrm{\vec{v}}$, donde $\mathrm{\vec{F}_{fricción}}$ es la fuerza de fricción, $\mathrm{\vec{v}}$ es la velocidad y $b$ es una constante positiva, las ecuaciones cambian del siguiente modo:

$$\ddot{x}=-b\dot{x}$$
$$\ddot{y}=-g-b\dot{y}$$

Así, si volvemos a definir a nuestras nuevas variables como:

$$x_{1}=x$$
$$x_{2}=\dot{x}$$
$$x_{3}=y$$
$$x_{4}=\dot{y}$$

El sistema de ecuaciones que debemos resolver para hallar la nueva trayectoria de la partícula en tiro parabólico que ahora está sometida a una fricción proporcional a la velocidad es la siguiente:

$$\dot{x_{1}}=x_{2}$$
$$\dot{x_{2}}=-bx_{2}$$
$$\dot{x_{3}}=x_{4}$$
$$\dot{x_{4}}=-g-bx_{4}$$

Por lo que la función con la que debemos alimentar a `euler_general` es $\mathbf{f}(\mathbf{x}(t),\,t)=(x_{2}(t),\,-bx_{2}(t),x_{4}(t),-g-bx_{4}(t))$, que será representada por la siguiente función:

```julia
tiroparabolico_friccionlineal(X,t,b=0.2)=[X[2],-b*X[2],X[4],-9.81-b*X[4]]
```

En el caso anterior, pusimos un valor por defecto al valor de $b$ como $b=0.1$. Es necesario que la función `tiroparabolico_friccionlineal` sólo tenga como únicos argumentos obligatorios al arreglo `X` y al escalar $t$ para que pueda constituir un argumento válido para `euler_general`, por el modo en que ésta última está programada. Eso pone un problema para poder variar a voluntad el valor de $b$ cuando queremos usar `euler_general`, ya que $b$ ya está fijo dentro de `tiroparabolico_friccionlineal`. Sin embargo, podemos crear una función (en el sentido `Function`) `tiroparabolico_friccionlinealB` que sólo dependa de $b$, y que nos regrese una `Function` basada en `tiroparabolico_friccionlineal` que siga tomando como únicos argumentos  al arreglo `X` y al escalar $t$ (para que pueda constituir un argumento válido para `euler_general`), pero ya con el valor de $b$ usado para alimentar a `tiroparabolico_friccionlinealB`. Implementamos a `tiroparabolico_friccionlinealB` como sigue:

```julia
function tiroparabolico_friccionlinealB(b)
    funct(X,t)=tiroparabolico_friccionlineal(X,t,b)
    return funct
end
```

Ahora, simplemente corroboramos que lo que entrega esta función son funciones:

```julia
typeof(tiroparabolico_friccionlinealB(0.3))
```

De ese modo, estamos listos para usar a `tiroparabolico_friccionlinealB` evaluada en cierto valor de $k$ como argumento de `euler_general`. Usamos el mismo valor para los otros argumentos (condición inicial e intervalo de tiempo) que en el inciso anterior para notar las diferencias:

```julia
T,X=euler_general(tiroparabolico_friccionlinealB(0.2),[20,30,30,70],0,15)
plot(X[1],X[3])
xlabel(L"$x(t)\,\mathrm{(m)}$")
ylabel(L"$y(t)\,\mathrm{(m)}$")
hlines(0, 0, 200, colors="k", linestyles="dashed")
vlines(0, -100, 250, colors="k", linestyles="dashed")
title("Tiro parabólico con fricción lineal")
```

Vemos inmediatamente que no sólo la forma de la curva es diferente a la obtenida en el Inciso (a), sino que además el alcance del proyectil fue bastante menor cuando se le sometió a la fuerza de fricción.

#### Inciso (c)

Ahora, si la fricción es proporcional al cuadrado de la rapidez, o sea, si es de la forma $\mathrm{\vec{F}_{fricción}}=-\gamma|\mathrm{\vec{v}}|^{2}\mathrm{\hat{v}}=-\gamma|\mathrm{\vec{v}}|\mathrm{\vec{v}}$, donde $\mathrm{\vec{F}_{fricción}}=-\gamma|\mathrm{\vec{v}}|\mathrm{\vec{v}}$ es igual a $-\gamma\sqrt{\dot{x}^{2}+\dot{y}^{2}}(\dot{x},\,\dot{y})=(-\gamma\dot{x}\sqrt{\dot{x}^{2}+\dot{y}^{2}},\,-\gamma\dot{y}\sqrt{\dot{x}^{2}+\dot{y}^{2}})$, tenemos entonces el siguiente sistema de ecuaciones para el movimiento de la partícula:

$$\ddot{x}=-\gamma\dot{x}\sqrt{\dot{x}^{2}+\dot{y}^{2}}$$
$$\ddot{y}=-g-\gamma\dot{y}\sqrt{\dot{x}^{2}+\dot{y}^{2}}$$

Si volvemos a definir la variables del sistema de ecuaciones de primer orden del modo en que lo hicimos en los incisos anteriores, ahora el sistema de ecuaciones para este problema queda del siguiente modo:

$$\dot{x_{1}}=x_{2}$$
$$\dot{x_{2}}=-\gamma x_{2}\sqrt{x_{2}^{2}+x_{4}^{2}}$$
$$\dot{x_{3}}=x_{4}$$
$$\dot{x_{4}}=-g-\gamma x_{4}\sqrt{x_{2}^{2}+x_{4}^{2}}$$

Por lo que la función con la que debemos alimentar a `euler_general` es $\mathbf{f}(\mathbf{x}(t),\,t)=(x_{2}(t),\,-\gamma x_{2}(t)\sqrt{x_{2}^{2}(t)+x_{4}^{2}(t)},x_{4}(t),-g-\gamma x_{4}(t)\sqrt{x_{2}^{2}(t)+x_{4}^{2}(t)})$, que será representada por la siguiente función:

```julia
tiroparabolico_friccioncuadratica(X,t,γ=0.02)=[X[2],-γ*X[2].*sqrt((X[2].^2)+(X[4].^2)),X[4],-9.81-γ*X[4].*sqrt((X[2].^2)+(X[4].^2))]
```

Luego, volvemos a definir una función que nos permita variar libremente el valor de $\gamma$, `tiroparabolico_friccioncuadraticaG`, que nos regresa una función basada en `tiroparabolico_friccioncuadratica` que use dicho valor y que siga siendo un argumento válido para `euler_general`:

#### Inciso (a)

Ahora creamos una función `euler_parabolico` que sea como `euler_general` pero que corra hasta que la partícula llegue a tierra y regrese sólo el tiempo de vuelo y el alcance. Para ello, ya no pedimos tiempo final, sólo tiempo inicial (que es necesario para la condición de frontera). De todos modos, definimos el tiempo final para que sea muy grande (de modo que en cualquier caso realista, la partícula toque el suelo antes de que se llegue a ese tiempo final). Ahora, como el arreglo de condiciones iniciales es $[x_{0},v_{x0},y_{0},v_{y0}]$ (porque las funciones de `tiro_parabolico` están escritas para aceptar el arreglo de condiciones inciales de esa forma), checamos que $y_{0}$ sea no negativo para que la función tenga sentido. Luego de llenar la primera columna de `Xs` (que corresponde a las posiciones y velocidades en el tiempo inicial), definimos la variable de altura, que nos dice qué tan encima del suelo está la partícula. Luego, implementamos el método de Euler, con la diferencia que en lugar de usar el `for` que llena columnas usamos un `while` que corre mientras la altura sea no negativa y el índice $n$ (inicializado en 2) sea menor que el número de columnas de `Xs`. Cada que termina de llenar una columna, checa el valor de $y$, que corresponde al renglón 3 de esa columna, y se lo asigna a la variable de altura, tras lo cual aumenta el índice n. El `while` se detiene si toda `Xs` es llena (poco probable en cualquier caso realista, pero es necesario poner ese alto para que el programa no intente llenar columnas que no existen) o si el valor de la altura se vuelve negativo, de modo que en ese momento deja de llenar a `Xs`. Como $n$ se actualiza después que la altura, la $n$ de interés es la anterior a la del punto donde la altura se volvió negativa. Por ende, la columna que nos importa es la $n-1$, donde $n$ es el valor que alcanzó el índice antes de detener el `while`. El tiempo de vuelo es simplemente el valor del arreglo de tiempos en esa columna menos el tiempo inicial, y el alcance es el valor de la posición horizontal en ese tiempo menos la posición inicial (restamos la posición inicial porque el alcance es una *distancia*, o sea, la diferencia entre la posición final y la inicial). Damos la salida como un arreglo en vez de una tupla para facilitar el inciso siguiente (donde debemos graficar ambas salida en función de otro parámetro).

```julia
function euler_parabolico(funcion_vectorial, arreglo_condiciones,tiempo_inicial)
num_var=length(arreglo_condiciones)
    if arreglo_condiciones[3]<0
        error("Altura inicial no válida")
    end
    h=0.001
    tiempo_final=tiempo_inicial+10000
    tiempos=[tiempo_inicial:h:tiempo_final]
    num_tiempos=length(tiempos)
    Xs=zeros(num_var,num_tiempos)
    for m in 1:num_var
        Xs[m,1]=arreglo_condiciones[m]
    end
    altura=Xs[3,1]
    n=2
    while altura>=0 && n<=num_tiempos
        for m in 1:num_var
            Xs[m,n]=Xs[m,n-1]+h*funcion_vectorial(columnan(Xs,n-1),tiempos[n-1])[m]
        end
        altura=Xs[3,n]
        n=n+1
    end
    tiempo_vuelo=tiempos[n-1]-tiempo_inicial
    alcance=Xs[1,n-1]-arreglo_condiciones[1]
    return [tiempo_vuelo, alcance]
end
```

Probamos la función con las condiciones iniciales que usamos en la Respuesta anterior para los tres tipos de tiro, para comparar el resultado (para el alcance) con el que aparece en las gráficas de aquella sección.

```julia
euler_parabolico(tiroparabolico,[20,30,30,70],0)
```

```julia
euler_parabolico(tiroparabolico_friccionlinealB(0.2),[20,30,30,70],0)
```

```julia
euler_parabolico(tiroparabolico_friccioncuadraticaG(0.02),[20,30,30,70],0)
```

Si regresamos a ver las gráficas, vemos que la intersección entre el eje $x$ y la gráfica de la trayectoria de la partícula es mayor al número que da nuestra función. Eso es porque la función, como ya dijimos regresa como alcance la distancia entre la posición donde la partícula toca el suelo y la posición inicial. Como aquí la posición inicial era x=20, la posición donde vuelva tocar el suelo es mayor, pero la *distancia* sí coincide con el resultado de nuestra función. Además, si usamos la gráfica interactiva de la sección anterior (ajustando las condiciones iniciales) y movemos el tiempo final, podemos ver a qué tiempo regresa cada partícula al suelo. El número sí coincide con el resultado de nuestra nueva función para el tiempo de vuelo (en ambos casos usamos tiempo inicial cero, pero aún si no, nuestra función lo calcularía adecuadamente).

Algo que es de notarse es que si ponemos velocidad inicial cero, nuestra función nos regresa el tiempo de caída de una partícula en caída libre (también con los tres tipos de fricción).

El oscilador armónico simple obedece la siguiente ecuación de movimiento:

$$\ddot{x}=-\omega^{2}x,$$

donde $\omega$ es la frecuencia de oscilación. 

Para resolverlo, definimos las siguientes variables del nuevo sistema de ecuaciones diferenciales de primer orden:

$$x_{1}=x$$
$$x_{2}=\dot{x}$$

Después, el sistema equivalente queda así:

$$\dot{x_{1}}=x_{2}$$
$$\dot{x_{2}}=-\omega^{2}x_{1}$$

Por lo que la función con la que debemos alimentar a `euler_general` es $\mathbf{f}(\mathbf{x}(t),\,t)=(x_{2}(t),\,-\omega^{2}x_{1}(t))$, que será representada por la siguiente función:

```julia
movimientoarmonicosimple(X,t,ω=1)=[X[2],-(ω^2)*X[1]]
```

Volvemos a crear la familia de funciones que dependen de $\omega:$

```julia
function movimientoarmonicosimpleF(ω)
    funct(X,t)=movimientoarmonicosimple(X,t,ω)
    return funct
end
```

Graficamos $x$ y $\dot{x}$ en función de $t$ para una $\omega$ dada y ciertas condiciones iniciales:

```julia
T,X=euler_general(movimientoarmonicosimpleF(1),[0,1],0,15)
plot(T,X[1])
xlabel(L"$t\,\mathrm{(s)}$")
ylabel(L"$x(t)\,\mathrm{(m)}$")
#hlines(0, 0, 200, colors="k", linestyles="dashed")
#vlines(0, -100, 250, colors="k", linestyles="dashed")
title("Movimiento armónico simple")
```
