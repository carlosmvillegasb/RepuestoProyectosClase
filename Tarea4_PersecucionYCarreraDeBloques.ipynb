### Respuesta 1.

Lo único que debemos hacer en este problema es implementar el método de Euler-Cromer con una función, y usarla para resolver el sistema de ecuaciones diferenciales para cada caso.

```julia
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v)
    q = Vector{Float64}[[q0]]
    v = Vector{Float64}[[v0]]
    t = Float64[t0]
    for ti in t0:h:tf
        v_nueva = v0 + h*f_v(ti,q0,v0)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
    end
    t,hcat(q...)',hcat(v...)'
end
```

Aunque para este ejercicio mejor usamos la función con condición de paro:

```julia
function euler_cromer_paro(q0,v0,t0,tf,h,f_q,f_v,paro)
    q = Vector{Float64}[]
    v = Vector{Float64}[]
    t = Float64[]
    for ti in t0:h:tf
        v_nueva = v0 + h*f_v(ti,q0,v0)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        
        if paro(ti,q0,v0) == true
            break
        end
    end
    t,hcat(q...)',hcat(v...)'
end
```

La velocidad del gato está dada por:

$$\vec{v}_{G}(t)=v_{G}(t)\frac{\vec{r}_{R}(t) - \vec{r}_{G}(t)}{|\vec{r}_{R}(t) - \vec{r}_{G}(t)|}$$

Donde $\vec{r}_{R}(t) - \vec{r}_{G}(t)=\left(x_{R}(t)-x_{G}(t),\,y_{R}(t)-y_{G}(t)\right)$, $|\vec{r}_{R}(t) - \vec{r}_{G}(t)|=\sqrt{\left(x_{R}(t)-x_{G}(t)\right)^{2}+\left(y_{R}(t)-y_{G}(t)\right)^{2}}$ y $v_{G}$ es constante en este problema.

Por ende, la velocidad del gato está dada por:

$$\vec{v}_{G}(t)=\frac{v_{G}}{\sqrt{\left(x_{R}(t)-x_{G}(t)\right)^{2}+\left(y_{R}(t)-y_{G}(t)\right)^{2}}}\left(x_{R}(t)-x_{G}(t),\,y_{R}(t)-y_{G}(t)\right)$$

Esta es la parte que debe incluirse en las funciones de velocidad que se implementarán en ambos casos. La velocidad del ratón es diferente en cada uno. 

Antes de definir las funciones vectoriales que nos definen la derivada de las velocidades `v` y posiciones `q`, debemos establecer la conversión sobre qué significan cada una de las entradas de sus salidas. La salida debe ser un arreglo de cuatro elementos tanto para la función de posiciones como para la de velocidades, ya que se tienen dos partículas (el gato y el ratón) en dos dimensiones. Las primeras dos entradas corresponderán al ratón y las últimas dos corresponderán al gato, en ambos casos dando primero la coordenada horizontal y luego la vertical.

De ese modo, implementando la velocidad del gato en ambos incisos, la función de la derivada de las posiciones es para ambos: 

```julia
function gatoyraton_q(t,q,v,vG=1)
[v[1],v[2],(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(q[1]-q[3]),(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(q[2]-q[4])]
end
```

```julia
function gatoyraton_qA(t,q,v)
    vG=sqrt((v[3]^2)+(v[4]^2))
[v[1],v[2],(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(q[1]-q[3]),(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(q[2]-q[4])]
end
```

Con propósitos de poder variar la rapidez del gato, como en la tarea pasada, la rapidez del gato es un parámetro opcional. Creamos una función que tome la rapidez del gato como argumento y nos devuelva una función cuyos argumentos sean sólo t,q y v:

```julia
function gatoyraton_q_vG(vG)
    funcq(t,q,v)=gatoyraton_q(t,q,v,vG)
    return funcq
end
```

También es importante saber la derivada de la velocidad del gato. Considerando que en ambos problemas su rapidez es constante, sólo debemos calcular la derivada de $\frac{v_{G}}{\sqrt{\left(x_{R}(t)-x_{G}(t)\right)^{2}+\left(y_{R}(t)-y_{G}(t)\right)^{2}}}\left(x_{R}(t)-x_{G}(t),\,y_{R}(t)-y_{G}(t)\right)$. De ese modo, la derivada de la velocidad del gato es:

$$\frac{d}{dt}\vec{v}_{G}(t)=\frac{v_{G}}{\sqrt{\left(x_{R}(t)-x_{G}(t)\right)^{2}+\left(y_{R}(t)-y_{G}(t)\right)^{2}}}\left(\dot{x}_{R}(t)-\dot{x}_{G}(t),\,\dot{y}_{R}(t)-\dot{y}_{G}(t)\right)-v_{G}\frac{\left(x_{R}(t)-x_{G}(t)\right)\left(\dot{x}_{R}(t)-\dot{x}_{G}(t)\right)+\left(y_{R}(t)-y_{G}(t)\right)\left(\dot{y}_{R}(t)-\dot{y}_{G}(t)\right)}{\left(\sqrt{\left(x_{R}(t)-x_{G}(t)\right)^{2}+\left(y_{R}(t)-y_{G}(t)\right)^{2}}\right)^{3}}\left(x_{R}(t)-x_{G}(t),\,y_{R}(t)-y_{G}(t)\right)$$

Por ende, las componentes de las funciones de las derivadas de la velocidad que corresponden al gato se ven así:

`(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(v[1]-v[3])-vG*(q[1]-q[3])*((q[1]-q[3])*(v[1]-v[3])+(q[2]-q[4])*(v[2]-v[4]))/((sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))^3),(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(v[2]-v[4])-vG*(q[2]-q[4])*((q[1]-q[3])*(v[1]-v[3])+(q[2]-q[4])*(v[2]-v[4]))/((sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))^3)`

De una vez es importante definir la condición de paro para el método de Euler Cromer en este problema. Esa se da cuando la distancia entre el gato y el ratón es menor que cierto valor de tolerancia:

```julia
function parogato(t,q,v)
    d=sqrt((q[1]-q[3])^2+(q[2]-q[4])^2)
    tolerancia=0.001
    if d<tolerancia
        b=true
    else
        b=false
    end
    return b
end
```

Como inicialmente el ratón está en el origen y el gato está en el punto $(x_{0},\,y_{0})$, el arreglo de condiciones iniciales para la posición, q0, puede definirse como una  función de $x_0$ y $y_0$, para poderlas variar a voluntad:

```julia
q0_inciso1(x0,y0)=[0,0,x0,y0]
```

Para las condiciones iniciales de la velocidad, el ratón inicialmente debe tener una componente horizontal de la velocidad igual a su rapidez `vR` y una componente vertical igual a cero. Para el gato, ambas componentes deben ser consistentes con la función que define a la velocidad del gato, la cual depende de la distancia entre los animales (que en el estado inicial sólo depende de la posición del gato inicial) y de la rapidez del gato. Así, podemos dar una función de la rapidez del ratón, así como de la rapidez y posición inicial del gato, para dar las condiciones iniciales de velocidad: 

```julia
function v0_inciso1(vR,x0=-1,y0=-1,vG=1)
    d=sqrt((x0^2)+(y0^2))
    [abs(vR),0,-x0*vG/d,-y0*vG/d]
end
```

Falta únicamente definir la función de la derivada de las velocidades para este inciso. Ya tenemos la derivada de la velocidad del gato, pero falta tener la derivada de la velocidad del ratón para que ésta sea un círculo. La posición del ratón es $\vec{r}_{R}(t)=\left(R_{0}\cos\left(\frac{v_{R}}{R_{0}}t\right),\,R_{0}\sin\left(\frac{v_{R}}{R_{0}}t\right)\right)$. Por ende, la derivada de la velocidad es: $\ddot{\vec{r}}_{R}=-\frac{v_{R}^{2}}{R_{0}^{2}}\vec{r}_{R}$. Así, definimos la función de la derivada de la velocidad de los animales no sólo en términos de la rapidez del gato, sino también del radio de la circunferencia y la rapidez del ratón:

```julia
function gatoyraton_v2(t,q,v,vG=1,vR=1,R0=1)
    [-((vR^2)/(R0^2))*q[1],-((vR^2)/(R0^2))*q[2],(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(v[1]-v[3])-vG*(q[1]-q[3])*((q[1]-q[3])*(v[1]-v[3])+(q[2]-q[4])*(v[2]-v[4]))/((sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))^3),(vG/sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))*(v[2]-v[4])-vG*(q[2]-q[4])*((q[1]-q[3])*(v[1]-v[3])+(q[2]-q[4])*(v[2]-v[4]))/((sqrt((q[1]-q[3])^2+(q[2]-q[4])^2))^3)]
end
```

```julia
function gatoyraton_v2_parametros(vG,vR,R0)
    funcv2(t,q,v)=gatoyraton_v2(t,q,v,vG,vR,R0)
    return funcv2
end
```

### Respuesta 2.

Tenemos tres variables $q$, $\chi$, $\rho$ y $\theta$ (donde $\chi=x/d$), junto con sus velocidades $v$. Por ende, tanto la función de derivadas de las posiciones como la función de derivadas de velocidades tienen salidas de tres componentes. Por convención, acordaremos que la primera sea para las derivadas relacionadas con $\rho$, la segunda para aquellas de $\chi$, y la última para las de $\theta$. La función de las derivadas de las posiciones siempre es esta:

```julia
function carrera_q(t,q,v)
    [v[1],v[2],v[3]]
end
```

En variables adimensionales, tenemos la constricción $\chi(\tau)=\rho(\tau)-2$, o bien $q_{2}(\tau)=q_{1}(\tau)-2$.

Mientras tanto, las derivadas de las velocidades son las siguientes:

$$\rho'' = \frac{1}{\mu+1}(\rho\theta'^2 + \cos\theta)$$
$$\chi''=\frac{1}{\mu+1}(\rho\theta'^2 + \cos\theta)$$
$$\theta'' = -\frac{1}{\rho}(2\rho'\theta' + \sin\theta)$$

O bien:

$$\dot{v}_{1} = \frac{1}{\mu+1}(q_{1}v_{3}^2 + \cos(q_{3}))\\$$
$$\dot{v}_{2} = \frac{1}{\mu+1}(q_{1}v_{3}^2 + \cos(q_{3}))\\$$
$$\dot{v}_{3} = -\frac{1}{q_{1}}(2v_{1}v_{3} + \sin(q_{3}))$$

Por ende, la función de las derivadas de las velocidades es:

```julia
function carrera_v(t,q,v,μ=1)
    [(1/(1+μ))*(q[1]*(v[3]^2)+cos(q[3])),(1/(1+μ))*(q[1]*(v[3]^2)+cos(q[3])),(-1/q[1])*((2*v[1]*v[3])+sin(q[3]))]
end
```

```julia
function carrera_v_M(μ)
    funcM(t,q,v)=carrera_v(t,q,v,μ)
    return funcM
end
```

#### Inciso 1.

Ahora, creamos funciones que nos den las condiciones iniciales de las posiciones y velocidades. Usamos la constricción del sistema al usar la condición inicial:$\chi_{0}=\rho_{0}-2$. 

```julia
function q0_carrera(ρ0,θ0)
    [ρ0,ρ0-2,θ0]
end
```

```julia
function v0_carrera(ν0,ω0)
    [ν0,ν0,ω0]
end
```

Podemos hacer una función basada en `euler_cromer` que calcule y regrese los tiempos (adimensionales y en segundos) a los que el bloque 1 llega a la polea y el bloque 2 toca la pared. Esta función simplemente realiza el método de Euler-Cromer hasta haber encontrado los tiempos $\tau_1$ y $\tau_2$ donde el bloque 1 llega a la polea y el bloque 2 toca la pared, respectivamente. Para ello, toma los mismos parámetros que `euler_cromer`, excepto el tiempo final. Inicializa las variables tau1 y tau2 en tiempos que son menores a t0. Luego, realiza el método de Euler Cromer en un ciclo while (en vez de un for) mientras ambas variables sigan siendo menores a t0. A cada tiempo ti dentro del ciclo, revisa que las entradas del arreglo q_nueva tengan los valores que deben tener cuando los bloques llegan a su meta. Cuando el bloque 1 llega a la polea, $\rho(\tau_{1})=2$ (q[1]=2) y $\chi(\tau_{1})=0$ (q[2]=0). Cuando el bloque 2 llega a la pared, simplemente $\theta(\tau_{1})=0$ (q[3]=0). Por ende, si a un tiempo ti las componentes de q_nueva cumplen esas condiciones, el programa asigna el valor de ti a la tau correspondiente. Cuando ambas tau han sido renombradas, ya ambas son mayores que t0, por lo que el ciclo while se detiene. Finalmente, calcula el tiempo en segundos usando $t=\tau\sqrt{\frac{d}{g}}$, donde d y g son argumentos opcionales de la función para la longitud $d$ de la cuerda y la aceleración de la gravedad $g$, cuyos valores por defecto son 1m (como en el enunciado que dice que la cuerda mide 2m, por lo que $d$ vale 1m) y $9.81\mathrm{m/s^{2}}$, respectivamente. Al final, lo que la función devuelve son un par de arreglos: Uno con las $\tau$ y el otro con las $t$ en segundos. 

```julia
function carrera_bloques(q0,v0,t0,h,f_q,f_v,d=1,g=9.81)
    tolerancia=h
    q = Vector{Float64}[[q0]]
    v = Vector{Float64}[[v0]]
    t = Float64[t0]
    tf=t0+1000
    tau1=t0-1000
    tau2=t0-1000
    lock1=0
    lock2=0
    ti=t0
    while (tau1<t0||tau2<t0)&&ti<=tf
        v_nueva = v0 + h*f_v(ti,q0,v0)
        q_nueva = q0 + h*f_q(ti,q0,v_nueva)
        if abs(q_nueva[1]-2)<=tolerancia&&lock1==0
            tau1=ti
            lock1=1
        end
        if abs(q_nueva[2])<=tolerancia&&lock1==0
            tau1=ti
            lock1=1
        end
        if abs(q_nueva[3])<=tolerancia&&lock2==0
            tau2=ti
            lock2=1
        end
        q0 = q_nueva
        v0 = v_nueva
        push!(t,ti)
        push!(q,[q0])
        push!(v,[v0])
        ti=ti+abs(h)
    end
    tiempo1=tau1*sqrt(d/g)
    tiempo2=tau2*sqrt(d/g)
    return [tau1,tau2],[tiempo1,tiempo2]
end
```

```julia
carrera_bloques(q0_carrera(1,pi/2),v0_carrera(0,0),0,0.0001,carrera_q,carrera_v_M(1))
```

**Por ende, con estas condiciones iniciales, si las masas de los bloques son iguales ($\mu=1$), el bloque 1 toca la polea antes de que el bloque 2 toque la pared. Lo hace en $\tau_{1}=2.2598$, que equivale a 0.721499 segundos si la cuerda mide 2m.**

Podemos comprobar esto gráficamente al graficar $\rho(\tau)$, $\theta(\tau)$ y $\chi(\tau)$ usando `subplot`. Además, dibujaremos en cada una de las gráficas una línea horizontal con los valores que deben alcanzar $\rho(\tau_1)$, $ \chi(\tau_1)$ y $\theta(\tau_2)$ cuando los bloques llegan a sus "metas". También volveremos a usar la función `carrera_bloques` para encontrar $\tau_1$ y $\tau_2$ y dibujar líneas verticales en esos valores. Si la intersección entre estas líneas dibujadas también intersecta a la gráfica, significa que $\tau_1$ y $\tau_2$ fueron calculadas correctamente. 

```julia
T,Q,V=euler_cromer(q0_carrera(1,pi/2),v0_carrera(0,0),0,3,0.0001,carrera_q,carrera_v_M(1))
tau,tiem=carrera_bloques(q0_carrera(1,pi/2),v0_carrera(0,0),0,0.0001,carrera_q,carrera_v_M(1))
subplots_adjust(hspace=0.0)
tau1=string(tau[1])[1:6]
tau2=string(tau[2])[1:6]
subplot(311)
ax1 = gca()
setp(ax1[:get_xticklabels](),visible=false)
ylabel(L"$\rho(\tau)$")
plot(T,Q[:,1],"r-")
hlines(2,T[1],T[length(T)],linestyles="dashed")
vlines(tau[1],Q[1,1],Q[size(Q,1),1],colors="c",linestyles="dashed",label=L"$\tau_{1}=$""$tau1")
axis("tight")
legend(loc=0)

subplot(312, sharex=ax1)
ax2 = gca()
setp(ax2[:get_xticklabels](),visible=false)
ylabel(L"$\chi(\tau)$")
xlabel(L"$t$")
plot(T,Q[:,2])
hlines(0,T[1],T[length(T)],linestyles="dashed")
vlines(tau[1],Q[1,2],Q[size(Q,1),2],colors="c",linestyles="dashed",label=L"$\tau_{1}=$""$tau1")
axis("tight")
legend(loc=0)

subplot(313, sharex=ax2)
ylabel(L"$\theta(\tau)$")
xlabel(L"$\tau$")
plot(T,Q[:,3],"g-")
hlines(0,T[1],T[length(T)],linestyles="dashed")
vlines(tau[2],Q[1,3],Q[size(Q,1),3],colors="m",linestyles="dashed",label=L"$\tau_{2}=$""$tau2")
axis("tight")
legend(loc=0)
```

#### Inciso 2.

Podemos hacer una función que nos permita graficar la posición del centro de masas (en metros) contra el tiempo. Simplemente debe tomar las condiciones iniciales y el intervalo de tiempo como argumentos, junto con el valor de $\mu$ y de $d$. Esos argumentos se los pasa al método de Euler-Cromer para hallar las posiciones, de las cuales depende el centro de masas.

```julia
function carrera_bloques_CentroMasa(q0,v0,t0,tf,h,μ,d=1)
    T,Q,V=euler_cromer(q0,v0,t0,tf,h,carrera_q,carrera_v_M(μ))
    Xcm=(d/(μ+1))*((Q[:,1].*(μ+sin(Q[:,3])))-2*μ)
    return T,Xcm
end
```

Graficamos la evolución del centro de masa en función de $\tau$. También vale la pena dibujar líneas verticales en los valores de $\tau$ para los cuales cada bloque llega a su "meta".

```julia
T,X=carrera_bloques_CentroMasa(q0_carrera(1,pi/2),v0_carrera(0,0),0,4,0.0001,1)
tau,tiem=carrera_bloques(q0_carrera(1,pi/2),v0_carrera(0,0),0,0.0001,carrera_q,carrera_v_M(1))
tau1=string(tau[1])[1:6]
tau2=string(tau[2])[1:6]
plot(T,X)
xlabel(L"$\tau$")
ylabel(L"$x_{\mathrm{cm}}\,\mathrm{(m)}$")
vlines(tau[1],X[1],X[size(X,1)],colors="r",linestyles="dashed",label=L"$\tau_{1}=$""$tau1")
vlines(tau[2],X[1],X[size(X,1)],colors="g",linestyles="dashed",label=L"$\tau_{2}=$""$tau2")
axis("tight")
legend(loc=0)
```

Como ambos bloques tienen la misma masa y están a la misma distancia horizontal del origen (la polea), la posición inicial del centro de masa es el origen. El hecho de que la posición del centro de masa sea positiva en todo momento significa que la distancia entre el bloque 1 y la polea es en todo momento menor a la distancia entre el bloque 2 y el plano que contiene a la pared. De ese modo, como el bloque 1 siempre está más cerca de su "meta", claramente será el primero en llegar a ella.

#### Inciso 3.

Podemos crear otra función que tome los mismos argumentos obligatorios que la anterior y regresa las energías (adimensionales) de los bloques y el sistema, para luego graficarlas en el mismo lienzo:

```julia
function carrera_bloques_Energia(q0,v0,t0,tf,h,μ,d=1,g=9.81)
    T,Q,V=euler_cromer(q0,v0,t0,tf,h,carrera_q,carrera_v_M(μ))
    k1=(μ/2)*V[:,1].*V[:,1]
    k2=(1/2)*((V[:,1].*V[:,1])+(Q[:,1].*V[:,3].*Q[:,1].*V[:,3]))
    u2=-Q[:,1].*cos(Q[:,3])
    e=k1+k2+u2
    return T,k1,k2,u2,e
end
```

```julia
T,K1,K2,U2,E=carrera_bloques_Energia(q0_carrera(1,pi/2),v0_carrera(0,0),0,4,0.0001,1)
plot(T,K1,label=L"$K_{1}/E_{0}$")
plot(T,K2,label=L"$K_{2}/E_{0}$")
plot(T,U2,label=L"$U_{2}/E_{0}$")
plot(T,E,label=L"$E/E_{0}$")
axis("tight")
xlabel(L"$\tau$")
ylabel("Energía")
legend(loc=0)
```

De ese modo, comprobamos que la energía del sistema se conserva.

Para encontrar a qué valor de $\mu$ se obtiene un empate entre ambos bloques (o sea $\tau_{1}=\tau_{2}$), hay que hacer una función de $\mu$ que devuelva $\tau_{1}-\tau_{2}$, i.e. $f(\mu)=\tau_{1}-\tau_{2}$. Después de eso, simplemente debe encontrarse la raíz con el método de bisección. Para comprobar gráficamente, simplemente debe graficarse $f(\mu)=\tau_{1}-\tau_{2}$ contra $\mu$:

```julia
function empate_bloques(μ)
    tau,tiem=carrera_bloques(q0_carrera(1,pi/2),v0_carrera(0,0),0,0.0001,carrera_q,carrera_v_M(μ))
    difencia_choques=tau[1]-tau[2]
    return difencia_choques
end
```

Usamos la función bisección de las tareas pasadas, y la evaluamos en nuestra función `empate_bloques`, dando como extremos del intervalo a $\mu_{a}=1$ y $\mu_{b}=2$:

```julia
function biseccion(f::Function,A::Real,B::Real)
    tolerancia=0.0001
    if abs(A-B)<=tolerancia
        error("intervalo no válido")
    elseif A>B
        a=B
        b=A
    else
        a=A
        b=B
    end
    if f(a)*f(b)>0
        error("No puede garantizarse que haya raíces")
        elseif abs(f(a))<=tolerancia
        c=a
        elseif abs(f(b))<=tolerancia
        c=b
    elseif f(a)*f(b)<0
        c=(a+b)/2
    end
    if abs(f(c))<=tolerancia
        c
        elseif f(a)*f(c)<0
        biseccion(f,a,c)
        elseif f(b)*f(c)<0
        biseccion(f,c,b)
    else
        error("error inesperado")
    end
end
```
