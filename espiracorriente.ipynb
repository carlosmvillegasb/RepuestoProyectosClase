```julia
import Elliptic
```

```julia
using PyPlot
```

```julia
function Bz(r,z)
    R=abs(r)
   a=1
    I=1
    μ0=1
    B0=I*μ0/(2*a)
    α=R/a
    β=z/a
    γ=z/r
    Q=((1+α)^2+β^2)
    H=Q-4*α
    k=(4*α/Q)
    if R==a && z==0
        Campo=0
    elseif R==0
        Campo=(μ0*I*(a^2))/(2*((a^2)+(z^2))^(3/2))
    else 
        Campo=(B0/(pi*sqrt(Q)))*(Elliptic.E(k)*((1-α^2-β^2)/H)+Elliptic.K(k))
    end
    return Campo
end
```

```julia
function Br(r,z)
    R=abs(r)
   a=1
    I=1
    μ0=1
    B0=I*μ0/(2*a)
    α=R/a
    β=z/a
    γ=z/r
    Q=((1+α)^2+β^2)
    H=Q-4*α
    k=(4*α/Q)
    if R==a && z==0
        Campo=0
        elseif R==0
        Campo=0
    else
        Campo=(B0*γ/(pi*sqrt(Q)))*(Elliptic.E(k)*((1+α^2+β^2)/H)-Elliptic.K(k))
    end
    return Campo
end
```

```julia
Z=[-3:0.25:3];
R=[-3:0.25:3];
```

```julia
Cr=[Br(j,i) for i in Z, j in R];
```

```julia
Cz=[Bz(j,i) for i in Z, j in R];
```

```julia
CNr=[Br(j,i)/sqrt((Br(j,i)^2)+(Bz(j,i)^2)) for i in Z, j in R];
```

```julia
CNz=[Bz(j,i)/sqrt((Br(j,i)^2)+(Bz(j,i)^2)) for i in Z, j in R];
```

```julia
quiver(R,Z,Cr,Cz,minshaft=1,minlength=0,headlength=5,color="k",pivot="middle")
hlines(0,-1,1,colors="r")
xlabel(L"$y$")
ylabel(L"$z$")
title("Campo magnético de una espira. Carlos Mauricio Villegas Burgos")
```
